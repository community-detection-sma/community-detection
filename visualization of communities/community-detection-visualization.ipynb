```python
import networkx as nx
import pandas as pd
from networkx.algorithms.community import greedy_modularity_communities
import matplotlib.pyplot as plt
```

```python
print("Jazz Musicians Network")
jazz = nx.read_weighted_edgelist("jazz.net")
nx.draw(jazz,with_labels = True)
```

```python
#Karate Club Network
#karate = nx.karate_club_graph()
# karate = nx.read_weighted_edgelist("karate.net")
karate = nx.read_gml("karate.gml",label = 'id')
nx.draw(karate,with_labels = True)
```

```python
#Dolphins Social Media Network

dolphins = nx.read_gml("dolphins.gml",label = 'id')
nx.draw(dolphins,with_labels = True)
```

```python
#for 2 communities
commJ = list(greedy_modularity_communities(jazz))
sorted(commJ[0])
sorted(commJ[1])
```

```python
# plot the communities
color_map = []
for node in jazz:
    if node in commJ[0]:
        color_map.append('blue')
    else: 
        color_map.append('green')
		
nx.draw(jazz,node_color=color_map, with_labels=True)
plt.show()
```

```python
#finding 3 communities
commK = list(greedy_modularity_communities(karate))
sorted(commK[0])
sorted(commK[1])
sorted(commK[2])
```

```python
# plot the communities
color_map = []
for node in karate:
    if node in commK[0]:
        color_map.append('blue')
    else: 
        if node in commK[1]:
            color_map.append('red')
        else: 
            color_map.append('purple')
		
nx.draw(karate,node_color=color_map, with_labels=True)
plt.show()
```

```python
commD = list(greedy_modularity_communities(dolphins))
sorted(commD[0])
sorted(commD[1])
sorted(commD[2])
sorted(commD[3])
```

```python
# plot the communities
color_map = []
for node in karate:
    if node in commK[0]:
        color_map.append('blue')
    else: 
        if node in commK[1]:
            color_map.append('red')
        else: 
            if node in commK[1]:
                color_map.append('green')
            else: 
                color_map.append('purple')
		
nx.draw(karate,node_color=color_map, with_labels=True)
plt.show()
```
